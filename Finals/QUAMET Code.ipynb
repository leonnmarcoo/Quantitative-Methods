## 1. Demographics Analysis

```python
# Demographics Analysis
# Assuming the survey has demographic questions - let's identify them

# Function to create beautiful visualizations with our color scheme
def create_demographic_chart(data, title, labels=None, chart_type='bar'):
    """Create demographic charts with consistent styling"""
    
    # Set up the figure with dark background
    fig, ax = plt.subplots(figsize=(10, 8), facecolor=bg_color)
    ax.set_facecolor(bg_color)
    
    if chart_type == 'bar':
        # Create bar chart
        bars = ax.bar(range(len(data)), data.values, color=color_palette[:len(data)])
        
        # Customize bar chart
        ax.set_xticks(range(len(data)))
        ax.set_xticklabels(data.index, fontsize=12, color='white')
        ax.set_ylabel('Count', fontsize=12, color='white')
        
        # Add value labels on bars
        for i, bar in enumerate(bars):
            height = bar.get_height()
            ax.text(bar.get_x() + bar.get_width()/2., height + 0.5,
                   f'{int(height)}', ha='center', va='bottom', fontsize=11, color='white')
        
        # Remove top and right spines
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['left'].set_color('white')
        ax.spines['bottom'].set_color('white')
        ax.tick_params(colors='white')
        
    elif chart_type == 'pie':
        # Create traditional pie chart (no donut hole)
        wedges, texts, autotexts = ax.pie(data.values, labels=data.index, autopct='%1.1f%%', 
                                         colors=color_palette[:len(data)], 
                                         textprops={'fontsize': 11, 'color': 'white'})
        
        # Note: Removed any center circle creation to ensure traditional pie chart
        
        # Style the text
        for text in texts:
            text.set_fontsize(12)
            text.set_color('white')
        for autotext in autotexts:
            autotext.set_fontsize(11)
            autotext.set_color('white')
            autotext.set_fontweight('bold')
    
    # Set title
    ax.set_title(title, fontsize=16, fontweight='bold', color='white', pad=20)
    
    # Remove axis for pie charts
    if chart_type == 'pie':
        ax.axis('equal')
    
    plt.tight_layout()
    plt.show()
    
    # Print summary statistics
    print(f"\nTotal responses: {data.sum()}")
    for category, count in data.items():
        percentage = (count / data.sum()) * 100
        print(f"{category}: {count} ({percentage:.1f}%)")
    print()

# Let's identify demographic columns by examining common patterns
demographic_keywords = ['age', 'gender', 'year', 'level', 'course', 'section']
demographic_cols = []

for col in df.columns:
    col_lower = col.lower()
    if any(keyword in col_lower for keyword in demographic_keywords):
        demographic_cols.append(col)

print("Potential demographic columns found:")
for col in demographic_cols:
    print(f"- {col}")
    
# If we can't find obvious demographic columns, let's look at the first few columns
# which typically contain demographics in surveys
if not demographic_cols:
    print("\nNo obvious demographic columns found. Examining first few columns:")
    demographic_cols = df.columns[:5].tolist()  # Assume first 5 columns are demographics

print("\nAnalyzing demographic data...")
```

```python
# Visualize Gender Distribution
if 'Gender' in df.columns:
    gender_counts = df['Gender'].value_counts()
    print("Gender Distribution:")
    print(gender_counts)
    create_demographic_chart(gender_counts, "Gender Distribution of CCIT Students", chart_type='pie')

# Visualize Year Level Distribution
if 'Year Level' in df.columns:
    year_counts = df['Year Level'].value_counts().sort_index()
    print("\nYear Level Distribution:")
    print(year_counts)
    create_demographic_chart(year_counts, "Year Level Distribution of CCIT Students", chart_type='bar')
```

# 📊 Survey Data Visualizations

This section contains comprehensive visualizations for each survey question, maintaining a consistent dark theme with #2C63EE color palette.

```python
# Enhanced Visualization Functions with Consistent Design
import matplotlib.patches as mpatches
from matplotlib.colors import LinearSegmentedColormap

# Create a custom colormap based on our primary color
def create_custom_colormap():
    colors = ['#1a1a2e', '#2C63EE', '#4D7AF0', '#6E91F2', '#8FA8F4', '#B0BFF6', '#D1D6F8', '#F3F5FD']
    return LinearSegmentedColormap.from_list('custom_blue', colors, N=256)

custom_cmap = create_custom_colormap()

def create_enhanced_chart(data, title, question_num="", chart_type='bar', figsize=(12, 8)):
    """
    Create beautiful, consistent charts with dark theme and blue color palette
    """
    fig, ax = plt.subplots(figsize=figsize)
    fig.patch.set_facecolor(bg_color)
    ax.set_facecolor(bg_color)
    
    # Prepare data
    if isinstance(data, pd.Series):
        if chart_type in ['bar', 'horizontal']:
            # Sort data for better visualization
            data = data.sort_values(ascending=(chart_type == 'horizontal'))
    
    if chart_type == 'bar':
        bars = ax.bar(range(len(data)), data.values, 
                     color=color_palette[:len(data)], 
                     edgecolor='white', linewidth=0.5, alpha=0.9)
        
        ax.set_xticks(range(len(data)))
        ax.set_xticklabels(data.index, rotation=45, ha='right', fontsize=10)
        
        # Add value labels on bars
        for i, (bar, value) in enumerate(zip(bars, data.values)):
            height = bar.get_height()
            ax.text(bar.get_x() + bar.get_width()/2, height + max(data.values)*0.01,
                   f'{value}', ha='center', va='bottom', color='white', 
                   fontweight='bold', fontsize=9)
        
        ax.set_ylabel('Number of Responses', fontsize=12, color='white')
        
    elif chart_type == 'horizontal':
        bars = ax.barh(range(len(data)), data.values, 
                      color=color_palette[:len(data)], 
                      edgecolor='white', linewidth=0.5, alpha=0.9)
        
        ax.set_yticks(range(len(data)))
        ax.set_yticklabels(data.index, fontsize=10)
        
        # Add value labels on bars
        for i, (bar, value) in enumerate(zip(bars, data.values)):
            width = bar.get_width()
            ax.text(width + max(data.values)*0.01, bar.get_y() + bar.get_height()/2,
                   f'{value}', ha='left', va='center', color='white', 
                   fontweight='bold', fontsize=9)
        
        ax.set_xlabel('Number of Responses', fontsize=12, color='white')
        
    elif chart_type == 'pie':
        wedges, texts, autotexts = ax.pie(data.values, labels=data.index, autopct='%1.1f%%', 
                                         colors=color_palette[:len(data)], startangle=90,
                                         textprops={'color': 'white', 'fontsize': 10})
        
        for autotext in autotexts:
            autotext.set_color('white')
            autotext.set_fontweight('bold')
            autotext.set_fontsize(9)
    
    elif chart_type == 'donut':
        wedges, texts, autotexts = ax.pie(data.values, labels=data.index, autopct='%1.1f%%', 
                                         colors=color_palette[:len(data)], startangle=90,
                                         pctdistance=0.85, 
                                         textprops={'color': 'white', 'fontsize': 10})
        
        # Draw circle in center to create donut effect
        centre_circle = plt.Circle((0,0), 0.70, fc=bg_color, linewidth=2, edgecolor='white')
        fig.gca().add_artist(centre_circle)
        
        for autotext in autotexts:
            autotext.set_color('white')
            autotext.set_fontweight('bold')
            autotext.set_fontsize(9)
    
    # Set title with question number
    full_title = f"{question_num}{title}" if question_num else title
    ax.set_title(full_title, fontsize=14, fontweight='bold', color='white', pad=20)
    
    # Add grid for bar charts
    if chart_type in ['bar', 'horizontal']:
        ax.grid(True, alpha=0.3, axis='y' if chart_type == 'bar' else 'x', color='#444444')
    
    # Style the axes
    ax.tick_params(colors='white', labelsize=10)
    for spine in ax.spines.values():
        spine.set_color('#444444')
        spine.set_linewidth(1)
    
    plt.tight_layout()
    plt.show()

def create_likert_chart(data, title, question_num="", figsize=(14, 8)):
    """
    Special function for Likert scale questions (rating/agreement scales)
    """
    fig, ax = plt.subplots(figsize=figsize)
    fig.patch.set_facecolor(bg_color)
    ax.set_facecolor(bg_color)
    
    # Sort data to show progression (e.g., Strongly Disagree to Strongly Agree)
    likert_order = ['Strongly Disagree', 'Disagree', 'Neutral', 'Agree', 'Strongly Agree']
    
    # Try to sort by likert scale if applicable
    try:
        ordered_data = data.reindex([item for item in likert_order if item in data.index])
        if len(ordered_data) < len(data):  # Add any missing categories
            remaining = data[~data.index.isin(likert_order)]
            ordered_data = pd.concat([ordered_data, remaining])
        data = ordered_data
    except:
        pass  # Use original order if sorting fails
    
    # Create diverging color palette for Likert scales
    likert_colors = ['#FF6B6B', '#FFA07A', '#FFD700', '#90EE90', '#32CD32']
    colors_to_use = likert_colors[:len(data)] if len(data) <= 5 else color_palette[:len(data)]
    
    bars = ax.bar(range(len(data)), data.values, 
                 color=colors_to_use, 
                 edgecolor='white', linewidth=1, alpha=0.9)
    
    ax.set_xticks(range(len(data)))
    ax.set_xticklabels(data.index, rotation=45, ha='right', fontsize=10)
    
    # Add value labels and percentages
    total_responses = data.sum()
    for i, (bar, value) in enumerate(zip(bars, data.values)):
        height = bar.get_height()
        percentage = (value / total_responses) * 100
        ax.text(bar.get_x() + bar.get_width()/2, height + max(data.values)*0.01,
               f'{value}\n({percentage:.1f}%)', ha='center', va='bottom', 
               color='white', fontweight='bold', fontsize=9)
    
    full_title = f"{question_num}{title}" if question_num else title
    ax.set_title(full_title, fontsize=14, fontweight='bold', color='white', pad=20)
    ax.set_ylabel('Number of Responses', fontsize=12, color='white')
    
    ax.grid(True, alpha=0.3, axis='y', color='#444444')
    ax.tick_params(colors='white', labelsize=10)
    
    for spine in ax.spines.values():
        spine.set_color('#444444')
        spine.set_linewidth(1)
    
    plt.tight_layout()
    plt.show()

print("Enhanced visualization functions created!")
```

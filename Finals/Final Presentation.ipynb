# **Popularity of Gaming Streaming Platforms among CCIT Students at National University-Manila**

## **Figure 1: Year Level**

```python
import matplotlib.pyplot as plt
import numpy as np

year_levels = ['1st Year', '2nd Year', '3rd Year', '4th Year', '5th Year or Above']
populations = [14, 14, 14, 10, 12]

fig, ax = plt.subplots(figsize=(10, 8))
fig.patch.set_facecolor('#171717')
ax.set_facecolor('#171717')

colors = ['#1E4A8C', '#2C63EE', '#4A7BF7', '#6B94FF', '#8FADFF']

def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return f'{val} ({pct:.1f}%)'
    return my_autopct

wedges, texts, autotexts = ax.pie(
    populations,
    labels = year_levels,
    autopct = make_autopct(populations),
    startangle = 90,
    colors = colors,
    textprops = {'color': 'white', 'fontsize': 11, 'fontweight': 'bold'},
)

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')
    autotext.set_fontsize(10)

ax.set_title(
    'Figure 1: Year Level',
    color = 'white',
    fontsize = 20,
    fontweight = 'bold',
    pad = 20
)

ax.axis('equal')

plt.tight_layout()
plt.show()
```

## **Figure 3: Which of the following gaming streaming platforms do you currently use? (Select all that apply)** 

```python
import matplotlib.pyplot as plt
import numpy as np

platforms = ['Twitch', 'YouTube', 'Facebook', 'TikTok', 'Discord', 'Trovo', 'Kick', 'Other']
usage_counts = [26, 35, 51, 36, 16, 1, 9, 1]

fig, ax = plt.subplots(figsize=(12, 8))
fig.patch.set_facecolor('#171717')
ax.set_facecolor('#171717')

colors = ['#1E4A8C', '#2C63EE', '#4A7BF7', '#6B94FF', '#8FADFF', '#A9C0FF', '#C5D4FF', '#E1ECFF']

bars = ax.bar(platforms, usage_counts, color=colors, edgecolor='white', linewidth=1)

for bar, value in zip(bars, usage_counts):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.5,
            f'{value}',
            ha='center', va='bottom', color='white', fontsize=11, fontweight='bold')

ax.set_title(
    'Figure 3: Which of the following gaming streaming platforms do you currently use? (Select all that apply)',
    color='white',
    fontsize=20,
    fontweight='bold',
    pad=20
)

ax.set_xlabel('Streaming Platforms', color='white', fontsize=14, fontweight='bold')
ax.set_ylabel('Number of Users', color='white', fontsize=14, fontweight='bold')

ax.tick_params(axis='x', colors='white', labelsize=10, rotation=45)
ax.tick_params(axis='y', colors='white', labelsize=10)

ax.set_ylim(0, max(usage_counts) + 5)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')

plt.tight_layout()
plt.show()
```

## **Figure 6: How often do you watch gaming streams on your primary platform?**

```python
import matplotlib.pyplot as plt
import numpy as np

frequency_labels = ['Never', 'Rarely', 'Sometimes', 'Often', 'Always']
percentages = [4.7, 20.3, 35.9, 23.4, 15.6]
frequency_counts = [round(64 * (p / 100)) for p in percentages]

fig, ax = plt.subplots(figsize=(14, 6))
fig.patch.set_facecolor('#171717')
ax.set_facecolor('#171717')

colors = ['#8FADFF', '#6B94FF', '#4A7BF7', '#2C63EE', '#1E4A8C']

y_pos = 0
bar_height = 0.6

cumulative_widths = np.cumsum([0] + percentages[:-1])

bars = []
for i, (label, percentage, cum_width) in enumerate(zip(frequency_labels, percentages, cumulative_widths)):
    bar = ax.barh(y_pos, percentage, left=cum_width, height=bar_height,
                  color=colors[i], edgecolor='white', linewidth=1, 
                  label=label)
    bars.append(bar)

for i, (label, percentage, cum_width, count) in enumerate(zip(frequency_labels, percentages, cumulative_widths, frequency_counts)):
    x_pos = cum_width + percentage / 2

    ax.text(x_pos, y_pos, f'{label}\n{count}\n({percentage}%)', 
                ha='center', va='center', color='white', 
                fontsize=9, fontweight='bold')

ax.set_title(
    'Figure 6: How often do you watch gaming streams on your primary platform?',
    color='white',
    fontsize=20,
    fontweight='bold',
    pad=20
)

ax.set_xlabel('Percentage (%)', color='white', fontsize=14, fontweight='bold')
ax.set_ylabel('')

ax.set_yticks([])
ax.set_ylim(-0.5, 0.5)

ax.set_xlim(0, 100)

ax.set_xticks([0, 20, 40, 60, 80, 100])
ax.set_xticklabels(['0%', '20%', '40%', '60%', '80%', '100%'])
ax.tick_params(axis='x', colors='white', labelsize=10)
ax.tick_params(axis='y', colors='white', labelsize=10)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('white')

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), 
          ncol=5, fancybox=True, shadow=True,
          facecolor='#2F2F2F', edgecolor='white',
          labelcolor='white')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

frequency_labels = ['Strongly disagree', 'Disagree', 'Neutral', 'Agree', 'Strongly agree']
percentages = [3.1, 4.7, 31.3, 29.7, 31.3]
frequency_counts = [round(64 * (p / 100)) for p in percentages]

fig, ax = plt.subplots(figsize=(14, 6))
fig.patch.set_facecolor('#171717')
ax.set_facecolor('#171717')

colors = ['#8FADFF', '#6B94FF', '#4A7BF7', '#2C63EE', '#1E4A8C']

y_pos = 0
bar_height = 0.6

cumulative_widths = np.cumsum([0] + percentages[:-1])

bars = []
for i, (label, percentage, cum_width) in enumerate(zip(frequency_labels, percentages, cumulative_widths)):
    bar = ax.barh(y_pos, percentage, left=cum_width, height=bar_height,
                  color=colors[i], edgecolor='white', linewidth=1, 
                  label=label)
    bars.append(bar)

for i, (label, percentage, cum_width, count) in enumerate(zip(frequency_labels, percentages, cumulative_widths, frequency_counts)):
    x_pos = cum_width + percentage / 2
    
    if percentage > 0:
        ax.text(x_pos, y_pos, f'{label}\n{count}\n({percentage}%)', 
                    ha='center', va='center', color='white', 
                    fontsize=9, fontweight='bold')

ax.set_title(
    'Figure 12: I would recommend my primary gaming streaming platform\nto my friends or classmates',
    color='white',
    fontsize=18,
    fontweight='bold',
    pad=20
)

ax.set_xlabel('Percentage (%)', color='white', fontsize=14, fontweight='bold')
ax.set_ylabel('')

ax.set_yticks([])
ax.set_ylim(-0.5, 0.5)

ax.set_xlim(0, 100)

ax.set_xticks([0, 20, 40, 60, 80, 100])
ax.set_xticklabels(['0%', '20%', '40%', '60%', '80%', '100%'])
ax.tick_params(axis='x', colors='white', labelsize=10)
ax.tick_params(axis='y', colors='white', labelsize=10)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('white')

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), 
          ncol=5, fancybox=True, shadow=True,
          facecolor='#2F2F2F', edgecolor='white',
          labelcolor='white')

plt.tight_layout()
plt.show()
```
